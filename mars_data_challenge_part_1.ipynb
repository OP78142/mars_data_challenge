```python
from bs4 import BeautifulSoup
import pandas as pd
from splinter import Browser
import time
# driver = webdriver.Chrome()
MarsNews_url = ' https://redplanetscience.com/'
```

```python
executable_path = {'executable_path':"C:/Program Files/Google/Chrome/Chromedriver.exe"}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Access Mars News
browser.visit(MarsNews_url)
time.sleep(2)

# Set HTML object
html = browser.html

# parse HTML using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

news_list = []


news_title = soup.find_all('div', class_="content_title")

news_tease = soup.find_all('div', class_='article_teaser_body')

for title in news_title:
    news_dict = {}
    print ("Title: "+title.text)
    news_dict['Title'] = title.text
    for tease in news_tease:
        print ("Teaser: "+tease.text)
        news_dict["Teaser"] = tease.text
        news_list.append(news_dict)
# news_tease = soup.find('div', class_='article_teaser_body').text


# print(news_title)
# print(news_tease)
```
